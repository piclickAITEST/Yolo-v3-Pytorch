```python
from PIL import Image
import json
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# JSON 파일 불러오기
with open('google_vision_api_annotation.json') as file:
    final_json = json.load(file)
    
print(len(final_json))
```

```python
cate_idx = {'Outer': 0,
            'Top': 0,
            'Bottom': 0,
            'Onepiece': 0,
            'Bag': 0,
            'Shoes' : 0,
            'Headwear' : 0,
            'Belt': 0,
            'Bracelet': 0,
            'Earring': 0,
            'Glasses': 0,
            'Necklace': 0,
            'Necktie': 0,
            'Scarf': 0,
            'Tiara': 0}
```

```python
Category = {'Outer' : ["Coat","Jacket","Outerwear","Suit"],
             'Top' : ["Shirt","Top"],
             'Bottom' : ["Jeans","Miniskirt","Pants","Shorts","Skirt","Trousers"],
             'Onepiece' : ["Dress","Overall"],
             'Bag' : ["Backpack","Bag","Briefcase","Handbag","Luggage & bags","Suitcase","Satchel","Tašky a zavazadla"],
             'Shoes' : ["Boot","Footwear","High-heeled footwear","Sandal","Shoe"],
             'Headwear' : ["Cowboy hat","Fedora",'Hat',"Straw hat","Sun hat"],
             'Belt':['Belt'],
             'Bracelet':['Bracelet'],
             'Earring':['Earring'],
             'Glasses':['Glasses', 'Sunglasses'],
             'Necklace':['Necklace'],
             'Necktie':['Necktie', 'Bowtie'],
             'Scarf':['Scarf'],
             'Tiara':['Tiara'],
            }

cate = []
for i in list(Category.values()):
    cate += i
print(len(cate))    
```

```python
# 카테고리에 해당하지 않은 박스에 대한 어노테이션 삭제

obj_list = []

for idx in range(0,len(final_json)):
    for obj in range(len(final_json[idx])-2):
        if not final_json[idx]['obj_'+str(obj)]['label'] in cate:
            obj_list.append(obj)
            #print('no!', final_json[idx]['obj_'+str(obj)]['label'])
        #else:
            #print('yes!', final_json[idx]['obj_'+str(obj)]['label'])
        
    if len(obj_list) == 0:
        print('no wrong box!')
    else:
        for o in obj_list:
            #print(obj_list)
            #print(o, '번째 obj 삭제')
            final_json[idx].pop('obj_'+str(o))
    obj_list = []       
```

```python
## bounding box가 없는 데이터 제거 process

none = []

for idx in range(0,len(final_json)):
    keys = []
    
    for key in final_json[idx].keys():
        keys.append(key)
    
    if len(keys) == 2:
        none.append(idx)
        
print(len(none),'개 bounding box 없음')   

#LIST INDEX DELETE : 역순서로 제거

for n in sorted(none, reverse=True):
    del final_json[n]
    
print('최종 이미지 수',len(final_json))
```

```python
# 라벨값 변경

for idx in range(0,len(final_json)):
    keys = []
    for key in final_json[idx].keys():
        keys.append(key)
    
    for obj in keys[2:]:
        for cate in Category.keys():
            if final_json[idx][obj]['label'] in Category[cate]:
                final_json[idx][obj]['label'] = str(cate)
            else:
                continue
```

```python
# label 값 15개의 라벨로 변경 작업
test = []

for idx in range(0,len(final_json)):
    keys = []
    for key in final_json[idx].keys():
        keys.append(key)
        
    for obj in keys[2:]:
        test.append(final_json[idx][obj]['label'])

# 변경 후 확인
print(len(test))
print(set(test))
```

```python
# 카테고리별 박스 갯수 파악

for idx in range(0,len(final_json)):
    keys = []
    for key in final_json[idx].keys():
        keys.append(key)
    
    if len(keys) == 2:
        print('no bounding box')
    for obj in keys[2:]:
        for c in cate_idx.keys():
            if final_json[idx][obj]['label'] == c:
                cate_idx[c] += 1
                
print(cate_idx)     

D = cate_idx

plt.figure(figsize=(16,16))
plt.bar(range(len(D)), list(D.values()), align='center')
plt.xticks(range(len(D)), list(D.keys()))
plt.show()
```

```python
# 특정 라벨 값의 이미지 check 시 사용

path = []
for idx in range(0,len(final_json)):
    keys = []
    for key in final_json[idx].keys():
        keys.append(key)
    for obj in keys[2:]:
        if final_json[idx][obj]['label'] == 'Tiara':
            path.append(final_json[idx]['file_path'])
```

```python
# 특정 이름의 라벨값 다시 idx로 변경

for idx in range(0,len(final_json)):
    keys = []
    for key in final_json[idx].keys():
        keys.append(key)
    
    for obj in keys[2:]:
        for c,t in enumerate(list(Category.keys())):
            if final_json[idx][obj]['label'] == t:
                final_json[idx][obj]['label'] = c
            else:
                continue
```
