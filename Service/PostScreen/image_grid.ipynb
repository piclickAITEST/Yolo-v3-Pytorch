```python
from PIL.ExifTags import TAGS
from PIL import ImageFilter, Image, ImageDraw

import glob
```

```python
img_list=[]
for path in glob.glob('../testimage/*'):
    img_list.append(path)
```

```python
idx = 2
grid_size_w = 8
grid_size_h = 4
```

```python
image = Image.open(img_list[idx])

draw = ImageDraw.Draw(image)
y_start = 0
y_end = image.height
step_size = int(image.width / grid_size_w)

for x in range(0, image.width, step_size):
    line = ((x, y_start), (x, y_end))
    draw.line(line, fill=255)

x_start = 0
x_end = image.width

step_size = int(image.height / grid_size_h)

for y in range(0, image.height, step_size):
    line = ((x_start, y), (x_end, y))
    draw.line(line, fill=255)

del draw
image
```
